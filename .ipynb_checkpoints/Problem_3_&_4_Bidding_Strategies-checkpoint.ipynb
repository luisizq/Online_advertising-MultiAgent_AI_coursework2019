# Linear and Non-linear Bidding Strategies

```python
import pandas as pd
import numpy as np
```

```python
validation = pd.read_csv('validation.csv')
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

## Linear Bidding

```python
def binary_search(arr, val):
    # returns index
    low, high = 0, len(arr) - 1
    
    while low < high:
        
        mid = int((low+high)/2)
        
        if arr[mid] < val:
            low = mid + 1
        else:
            high = mid
    
    return low

def linear_bidding(dat,base_bids):
    
    results = []
    
    for b in base_bids:
        bids = pd.Series(b * ((pred / avgCTR)))
        budget = np.cumsum(dat.payprice*(dat.payprice <= bids))
        fin = binary_search(budget, limited_budget)
        n_clk = np.sum((dat.click*(dat.payprice <= bids))[:fin]) 
        n_im = np.sum((dat.payprice <= bids)[:fin])
        cost = budget[fin]

        # labels = ['bid', 'clicks', 'imps_won', 'total_spend']
        results.append( [b, n_clk, n_im, cost] )

    return results

def calculating_linear_bids(model):
    if model=='logreg':
        df=pd.read_csv('pctr_val_logreg.csv')
    elif model == 'rf':
        df=pd.read_csv('pctr_val_rf.csv')
 
    pred=list(df['pctr'])
    base_bids = np.arange(1,502,1)
    a = linear_bidding(validation,base_bids)
    linear_bids=pd.DataFrame.from_records(a,columns=["base_bids","clicks","impressions","cost"])
    linear_bids['CTR'] = (linear_bids.clicks / linear_bids.impressions * 100).round(4).astype(str)
    linear_bids['CPM'] = (linear_bids.cost / linear_bids.impressions * 1000).round(2).astype(str)
    linear_bids['CPC'] = (linear_bids.cost / linear_bids.clicks).round(2).astype(str)
    linear_bids.to_csv('linear_bidding_prices_'+model+'.csv',index=False)
    print(linear_bids.iloc[np.where(linear_bids.clicks == linear_bids.clicks.max())])

limited_budget=6250000
avgCTR = (train.click.sum() / len(train))
calculating_linear_bids('logreg')
calculating_linear_bids('rf')
```

## Square Bidding

```python
def square_bidding(dat,base_bids):
    
    results = []
    
    for b in base_bids:
        bids = pd.Series(b * ((pred / avgCTR)**2))
        budget = np.cumsum(dat.payprice*(dat.payprice <= bids))
        fin = binary_search(budget, limited_budget)
        n_clk = np.sum((dat.click*(dat.payprice <= bids))[:fin]) 
        n_im = np.sum((dat.payprice <= bids)[:fin])
        cost = budget[fin]

        # labels = ['bid', 'clicks', 'imps_won', 'total_spend']
        results.append( [b, n_clk, n_im, cost] )

    return results

def calculating_sqr_bids(model):
    if model=='logreg':
        df=pd.read_csv('pctr_val_logreg.csv')
    elif model == 'rf':
        df=pd.read_csv('pctr_val_rf.csv')
    
    pred=list(df['pctr'])
    base_bids = np.arange(1,502,1)
    a = square_bidding(validation,base_bids)
    square_bids=pd.DataFrame.from_records(a,columns=["base_bids","clicks","impressions","cost"])
    square_bids['CTR'] = (square_bids.clicks / square_bids.impressions * 100).round(4).astype(str)
    square_bids['CPM'] = (square_bids.cost / square_bids.impressions * 1000).round(2).astype(str)
    square_bids['CPC'] = (square_bids.cost / square_bids.clicks).round(2).astype(str)
    print(square_bids.iloc[np.where(square_bids.clicks == square_bids.clicks.max())])
    square_bids.to_csv('sqaure_bidding_prices_'+model+'.csv',index=False)

limited_budget=6250000
avgCTR = (train.click.sum() / len(train))
calculating_sqr_bids('logreg')
calculating_sqr_bids('rf')
```

## Exponential Bidding

```python
def exp_bidding(dat,base_bids):
    
    results = []
    
    for b in base_bids:
        bids = pd.Series(b * np.exp(pred / avgCTR))
        budget = np.cumsum(dat.payprice*(dat.payprice <= bids))
        fin = binary_search(budget, limited_budget)
        n_clk = np.sum((dat.click*(dat.payprice <= bids))[:fin]) 
        n_im = np.sum((dat.payprice <= bids)[:fin])
        cost = budget[fin]

        # labels = ['bid', 'clicks', 'imps_won', 'total_spend']
        results.append( [b, n_clk, n_im, cost] )

    return results

def calculating_exp_bids(model):
    if model=='logreg':
        df=pd.read_csv('pctr_val_logreg.csv')
    elif model == 'rf':
        df=pd.read_csv('pctr_val_rf.csv')

    pred=list(df['pctr'])
    base_bids = np.arange(1,502,1)
    a = exp_bidding(validation,base_bids)
    exp_bids=pd.DataFrame.from_records(a,columns=["base_bids","clicks","impressions","cost"])
    exp_bids['CTR'] = (exp_bids.clicks / exp_bids.impressions * 100).round(4).astype(str)
    exp_bids['CPM'] = (exp_bids.cost / exp_bids.impressions * 1000).round(2).astype(str)
    exp_bids['CPC'] = (exp_bids.cost / exp_bids.clicks).round(2).astype(str)
    print(exp_bids.iloc[np.where(exp_bids.clicks == exp_bids.clicks.max())])
    exp_bids.to_csv('exp_bidding_prices_'+model+'.csv',index=False)

limited_budget=6250000
avgCTR = (train.click.sum() / len(train))
calculating_exp_bids('logreg')
calculating_exp_bids('rf')
```
